```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# 1. Carregar o arquivo
df = pd.read_csv('rankingok.csv')

# --- CORREÇÃO AQUI ---
# 2. Converter a coluna 'data' para datetime ANTES de calcular o mínimo
df['data'] = pd.to_datetime(df['data'])

# 3. Calcular a data inicial (agora como um objeto datetime)
data_inicial = df['data'].min()

# 4. Criar a variável 'dias_desde_inicio' (número de dias)
df['dias_desde_inicio'] = (df['data'] - data_inicial).dt.days

# --- REGRESSÃO LINEAR (Prever nota com base na posição) ---

# 5. Definir X e Y (Variáveis preditora e dependente)
# X: posicao (Preditora)
# Y: nota (Dependente)
X = df[['posicao']]
Y = df['nota']

# 6. Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(X, Y)

# 7. Exibir resultados
coeficiente = modelo.coef_[0]
intercepto = modelo.intercept_
r_quadrado = modelo.score(X, Y)

print("--- Resultados da Regressão Linear Simples ---")
print(f"Variável Independente (X): posicao")
print(f"Variável Dependente (Y): nota")
print("-" * 40)
print(f"Coeficiente (Inclinação): {coeficiente:.4f}")
print(f"Intercepto: {intercepto:.4f}")
print(f"R-quadrado (Score do Modelo): {r_quadrado:.4f}")

# Exemplo de Previsão: Qual a nota esperada para a posicao 5?
posicao_previsao = 5
predicao = modelo.predict(np.array([[posicao_previsao]]))
print(f"\nPrevisão de Nota para Posição {posicao_previsao}: {predicao[0]:.4f}")
```
